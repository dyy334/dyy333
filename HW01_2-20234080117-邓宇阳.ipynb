```python
import numpy as np
np.random.seed(42)
```

```python
X = np.random.randn(3, 4)
print("X =")
print(X)
```

```python
Y = np.ones((4, 2))
print("Y =")
print(Y)
```

```python
Z = X @ Y
print("Z =")
print(Z)
```

```python
element_row1_col2 = Z[0, 1]
row2_all = Z[1, :]
print(f"第一行第二列: {element_row1_col2}")
print(f"第二行: {row2_all}")
```

```python
fro_norm = np.linalg.norm(Z, 'fro')
print(f"Frobenius 范数: {fro_norm}")
```
